```python
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import torch
import os 
import numpy as np
```

```python
# run in 3 minutes
model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
```

```python
image = Image.open('/Users/VoThinhPhat/Desktop/chatKPT-2024-AIC-HCMC/data/batch1/keyframes/keyframes_L01/L01_V001/001.jpg')
inputs = processor(images=image, return_tensors="pt")
with torch.no_grad():
    image_embedding = model.get_image_features(**inputs)
print("Image Embedding:", image_embedding)
```

```python
# run in about 2 minutes
embeddings_list = []
folder_path = "/Users/VoThinhPhat/Desktop/chatKPT-2024-AIC-HCMC/data/batch1/keyframes/keyframes_L01/L01_V001"

for image_name in os.listdir(folder_path):
    image_path = os.path.join(folder_path, image_name)
    if image_path.endswith(('.png', '.jpg', '.jpeg')):
        image = Image.open(image_path)
        inputs = processor(images=image, return_tensors="pt")
        with torch.no_grad():
            image_embedding = model.get_image_features(**inputs)
        embeddings_list.append(image_embedding.cpu().numpy().squeeze())

embeddings_array = np.array(embeddings_list)
np.save('image_embeddings.npy', embeddings_array)
```
