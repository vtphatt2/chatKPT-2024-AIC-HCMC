```python
import paramiko
import os
```

```python
def download_file(remote_host, username, password, remote_path, local_path):
    try:
        transport = paramiko.Transport((remote_host, 22))
        transport.connect(username=username, password=password)

        sftp = paramiko.SFTPClient.from_transport(transport)
        sftp.get(remote_path, local_path)

        sftp.close()
        transport.close()
        print(f"File downloaded successfully to {local_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

def check_remote_file_exists(remote_host, username, password, remote_path):
    try:
        ssh = paramiko.SSHClient()
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        ssh.connect(remote_host, username=username, password=password)
        command = f"test -f {remote_path} && echo 'File exists' || echo 'File does not exist'"
        stdin, stdout, stderr = ssh.exec_command(command)
        result = stdout.read().decode().strip()
        ssh.close()

        if result == "File exists":
            return True
        else:
            return False

    except Exception as e:
        print(f"An error occurred: {e}")
        return False

def download_files_with_prefix(remote_host, username, password, remote_directory, prefix, local_directory):
    try:
        transport = paramiko.Transport((remote_host, 22))
        transport.connect(username=username, password=password)
        sftp = paramiko.SFTPClient.from_transport(transport)
        files = sftp.listdir(remote_directory)
        matching_files = [file for file in files if file.startswith(prefix)]
        os.makedirs(local_directory, exist_ok=True)
        for file in matching_files:
            remote_file_path = os.path.join(remote_directory, file)
            local_file_path = os.path.join(local_directory, file)
            sftp.get(remote_file_path, local_file_path)
            print(f"Downloaded: {file}")

        sftp.close()
        transport.close()
        print("All matching files downloaded successfully.")

    except Exception as e:
        print(f"An error occurred: {e}")
```
